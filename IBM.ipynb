## 5.1 Relationship between Attrition rate and Job Role <a id=5.1></a>

```python
Attrition_job_role = ((ibm[ibm['Attrition'] == 'Yes'].groupby('JobRole').count()/ibm.groupby('JobRole').count())['Age'] * 100).reset_index()

Attrition_job_role.rename(columns = {'Age':'Attrition_rate'},inplace = True)

Attrition_job_role.sort_values('Attrition_rate', ascending = False, inplace = True)

Attrition_job_role
```

```python
Attrition_job_role['Attrition_rate'].round(2)
```

```python
sns.barplot(data = Attrition_job_role, y = 'JobRole', x = 'Attrition_rate', orient = 'h')

plt.title("Atrrition rate by Job Role")

plt.annotate(text = '39.76%',
             xy = (34,0),verticalalignment = 'center')
plt.annotate(text = '23.94%',
             xy = (24,1),verticalalignment = 'center')
plt.annotate(text = '23.08%',
             xy = (24,2),verticalalignment = 'center')
plt.annotate(text = '17.48%',
             xy = (18,3),verticalalignment = 'center')
plt.annotate(text = '16.10%',
             xy = (17,4),verticalalignment = 'center')
plt.annotate(text = '6.90%',
             xy = (7,5),verticalalignment = 'center')
plt.annotate(text = '6.87%',
             xy = (7,6),verticalalignment = 'center')
plt.annotate(text = '4.90%',
             xy = (5,7),verticalalignment = 'center')
plt.annotate(text = '2.50%',
             xy = (3,8),verticalalignment = 'center')


plt.tight_layout()

plt.show()
```

## 5.2 Relationship between Attrition rate and DistanceFromHome <a id=5.2></a>

```python
ibm.head()
```

```python
ibm_DistanceFromHome = ibm.copy()

ibm_DistanceFromHome['Attrition'].replace({'Yes':1,'No':0},inplace = True)

ibm_DistanceFromHome.head()
```

```python
ibm_DistanceFromHome['DistanceFromHome'].min()
```

```python
ibm_DistanceFromHome['DistanceFromHome'].max()
```

```python
distance_group = [(1,5),(6,10),(11,15),(16,20),(21,25),(25,30)]

def assign_dist_group(distance):
    for dist in distance_group:
        if dist[0] <= distance <= dist[1]:
            return f"{dist[0]} - {dist[1]}"
    return "unknow"

ibm_DistanceFromHome['Distance_range'] = ibm_DistanceFromHome['DistanceFromHome'].apply(assign_dist_group)
```

```python
sns.regplot(data = ibm_DistanceFromHome, y = 'Attrition', x = 'DistanceFromHome')

plt.ylabel('Attrition Rate')

plt.plot()
```

```python
sns.violinplot(data = ibm, x = 'Attrition', y = 'DistanceFromHome')

plt.show()
```

```python
Distance_Attrition = ibm_DistanceFromHome.groupby('Distance_range')['Attrition'].mean().reset_index()

Distance_Attrition.rename(columns = {'Attrition':'Attrition_rate'},inplace = True)

Distance_Attrition
```

```python
distance_order = ['1 - 5','6 - 10','11 - 15','16 - 20','21 - 25','25 - 30']

sns.barplot(data = Distance_Attrition, x = 'Distance_range', y = 'Attrition_rate',order = distance_order)

plt.show()
```
